```python
weights={
    
    'Food&Drinks':0.10,
    'AlcoholicItems':0.02,
    'Cloothing&Footwear':0.025,
    'Fuel&Power':0.10,
    'HouseholdGoods':0.06,
    'Health':0.010,
    'Transport':0.10,
    'Communication':0.02,
    'Recreation&Culture':0.07,
    'Education':0.01, 
    'Restaurants&Hotels':0.05,
    'Goods&Services':0.05,
    'AllExpenditure':1,
    'OtherExpenditure':0.10,
    'Total Exp':1,
    'Income':1
}
```

```python
df1["Rating"]=np.nan
```

```python
len(df1.columns)
```

```python
for i in range(len(df1)):
   
    c=0
    ratings=[0]*13
    for j in df1.columns:
        if j=="Rating" or j=='AllExpenditure' or j=='Total Exp' or j=='Income':
            continue
        ideal_bar=round(df1.loc[i,"Income"]*weights[j],2)
        val=df1.loc[i,j]
        if val<=ideal_bar:
            ratings[c]=1
        else:
            
            if val<0.10*ideal_bar+ideal_bar:
                 ratings[c]=2
            
            elif val<0.20*ideal_bar+ideal_bar:
                 ratings[c]=3

            elif val<0.30*ideal_bar+ideal_bar:
                 ratings[c]=4

            elif val<0.40*ideal_bar+ideal_bar:
                 ratings[c]=5

            elif val<0.50*ideal_bar+ideal_bar:
                 ratings[c]=6

            elif val<0.60*ideal_bar+ideal_bar:
                 ratings[c]=7

            elif val<0.70*ideal_bar+ideal_bar:
                 ratings[c]=8

            elif val<0.80*ideal_bar+ideal_bar:
                 ratings[c]=9
                
            elif val<ideal_bar+ideal_bar:
                 ratings[c]=10
            
            else:
                 ratings[c]=10
        c+=1
    
    print(np.mean(ratings))
    
    df1.loc[i,"Rating"]=round(np.mean(ratings),0)
    
```
